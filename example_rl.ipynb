```python
from demo.reinforcement_learning.actor_critic import ActorCritic
from demo.contextual_bandit_dgp import Simple
import numpy as np
from itertools import count
```

```python
from itertools import count

# Number of samples
n_samples = 1000

# Initialize the DGP
n_arms = 3
n_features = 5
shared_weight = 0.3
assigned_arms = np.random.randint(0, n_arms, size=n_samples)

dgp = Simple(n_arms=n_arms, n_features=n_features, shared_weight=shared_weight)

# Initialize Actor-Critic
actor_critic = ActorCritic(n_features=n_features)
running_reward = 10
for i_episode in count(1):
    ep_reward = 0
    for episode in range(100):
        state = dgp.x(num_units=1).reshape(1, -1)
        #select action from policy
        action = actor_critic.select_action(state)
        reward = dgp.r(a=action, x=state)
        actor_critic.rewards.append(reward)
        ep_reward += reward

    # update cumulative reward
    running_reward = 0.05 * ep_reward + (1 - 0.05) * running_reward

    # perform backprop
    actor_critic.finish_episode()

    # log results
    if i_episode % 10 == 0:
        print(f'Episode {i_episode}\tLast reward: {float(ep_reward):.2f}\tAverage reward: {float(running_reward):.2f}')

    if running_reward > 60:
        break

```
